```python
create_orders_table = """
CREATE TABLE IF NOT EXISTS orders1 (
    order_id INT PRIMARY KEY,
    order_date DATE,
    ship_mode VARCHAR(50),
    segment VARCHAR(50),
    country VARCHAR(100),
    city VARCHAR(100),
    state VARCHAR(100),
    postal_code VARCHAR(20),
    region VARCHAR(50)
);
"""

create_order_items_table = """
CREATE TABLE IF NOT EXISTS order_items (
    order_id INT,
    product_id VARCHAR(50),
    category VARCHAR(100),
    sub_category VARCHAR(100),
    cost_price FLOAT,
    list_price FLOAT,
    quantity INT,
    discount_percent FLOAT,
    discount_amount FLOAT,
    sale_price FLOAT,
    profit FLOAT,
    FOREIGN KEY (order_id) REFERENCES orders1(order_id) ON DELETE CASCADE
);
"""

mycursor.execute(create_orders_table)
mycursor.execute(create_order_items_table)
conn.commit()
print("Tables created successfully!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import mysql.connector

# Database connection
conn = mysql.connector.connect(
    host="gateway01.ap-southeast-1.prod.aws.tidbcloud.com",
    port=4000,
    user="48vkBMiRYz6FR8V.root",
    password="n74mpTyAKSdSMTOh",
    database="PRO_1",
)
mycursor = conn.cursor()

# Streamlit app
st.title("Business Insights Dashboard")

# Dropdown to select a query
query_options = {
    "1. Top 10 Highest Revenue Generating Products": """
        SELECT product_id, SUM(sale_price) AS total_revenue
        FROM order_items
        GROUP BY product_id
        ORDER BY total_revenue DESC
        LIMIT 10;
    """,
    "2. Top 5 Cities with the Highest Profit Margins": """
        SELECT o.city, SUM(oi.profit) AS total_profit
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY o.city
        ORDER BY total_profit DESC
        LIMIT 5;
    """,
    "3. Total Discount Given for Each Category": """
        SELECT category, SUM(discount_amount) AS total_discount
        FROM order_items
        GROUP BY category;
    """,
    "4. Average Sale Price per Product Category": """
        SELECT category, AVG(sale_price) AS avg_sale_price
        FROM order_items
        GROUP BY category;
    """,
    "5. Region with the Highest Average Sale Price": """
        SELECT o.region, AVG(oi.sale_price) AS avg_sale_price
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY o.region
        ORDER BY avg_sale_price DESC
        LIMIT 1;
    """,
    "6. Total Profit per Category": """
        SELECT category, SUM(profit) AS total_profit
        FROM order_items
        GROUP BY category;
    """,
    "7. Top 3 Segments with the Highest Quantity of Orders": """
        SELECT segment, SUM(quantity) AS total_quantity
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY segment
        ORDER BY total_quantity DESC
        LIMIT 3;
    """,
    "8. Average Discount Percentage Given per Region": """
        SELECT o.region, AVG(oi.discount_percent) AS avg_discount_percent
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY o.region;
    """,
    "9. Product Category with the Highest Total Profit": """
        SELECT category, SUM(profit) AS total_profit
        FROM order_items
        GROUP BY category
        ORDER BY total_profit DESC
        LIMIT 1;
    """,
    "10. Total Revenue Generated per Year": """
        SELECT YEAR(o.order_date) AS order_year, SUM(oi.sale_price) AS total_revenue
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY order_year;
    """,
    "11. Top 5 Products with the Highest Profit Margins": """
        SELECT product_id, SUM(profit) AS total_profit
        FROM order_items
        GROUP BY product_id
        ORDER BY total_profit DESC
        LIMIT 5;
    """,
    "12. Month with the Highest Sales": """
        SELECT MONTH(o.order_date) AS order_month, SUM(oi.sale_price) AS total_sales
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY order_month
        ORDER BY total_sales DESC
        LIMIT 1;
    """,
    "13. Top 3 Sub-Categories with the Highest Revenue": """
        SELECT sub_category, SUM(sale_price) AS total_revenue
        FROM order_items
        GROUP BY sub_category
        ORDER BY total_revenue DESC
        LIMIT 3;
    """,
    "14. Average Profit Margin per Product": """
        SELECT product_id, AVG((profit / sale_price) * 100) AS avg_profit_margin
        FROM order_items
        GROUP BY product_id;
    """,
    "15. Total Number of Orders per Region": """
        SELECT region, COUNT(order_id) AS total_orders
        FROM orders1
        GROUP BY region;
    """,
    "16. Top 5 Customers with the Highest Spending": """
        SELECT o.customer_id, SUM(oi.sale_price) AS total_spending
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY o.customer_id
        ORDER BY total_spending DESC
        LIMIT 5;
    """,
    "17. Most Popular Ship Mode": """
        SELECT ship_mode, COUNT(order_id) AS total_orders
        FROM orders1
        GROUP BY ship_mode
        ORDER BY total_orders DESC
        LIMIT 1;
    """,
    "18. Total Revenue per Segment": """
        SELECT segment, SUM(oi.sale_price) AS total_revenue
        FROM orders1 o
        JOIN order_items oi ON o.order_id = oi.order_id
        GROUP BY segment;
    """,
    "19. Top 3 Products with the Highest Discounts": """
        SELECT product_id, SUM(discount_amount) AS total_discount
        FROM order_items
        GROUP BY product_id
        ORDER BY total_discount DESC
        LIMIT 3;
    """,
    "20. Average Quantity Ordered per Product": """
        SELECT product_id, AVG(quantity) AS avg_quantity
        FROM order_items
        GROUP BY product_id;
    """,
}

# Dropdown to select a query
selected_query = st.selectbox("Select a Query", list(query_options.keys()))

# Execute the selected query
query = query_options[selected_query]
mycursor.execute(query)
result = mycursor.fetchall()
columns = [col[0] for col in mycursor.description]
df = pd.DataFrame(result, columns=columns)

# Display the results
st.header(selected_query)
st.dataframe(df)

# Close the database connection
conn.close()
```
